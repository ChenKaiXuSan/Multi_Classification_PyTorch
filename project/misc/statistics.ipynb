```python
import os 

disease_type = ['ASD', 'ASD_not']
flag = ['train', 'val']

path = '/workspace/data/segmentation_dataset_512/fold0/'
all_path = []

for f in flag:
    for d in disease_type:
        for l in os.listdir(os.path.join(path, f, d)):
            all_path.append(l)
        
```

```python
all_path.sort()
all_path
```

```python
four_disease = ['ASD', 'DHS', 'LCS', 'HipOA']

ASD_dict = {}
DHS_dict = {}
LCS_dict = {}
HipOA = {}
path_dict = {i:[] for i in four_disease}

def find_para(char):
    if char in four_disease:
        return char


for p in all_path:
    for i in p.split('_'):
        if i in four_disease:
            path_dict[i].append(p)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_pie(data: dict, dis_type):
    y = np.array([35, 25, 25, 15])

    plt.pie(data.values(),
            labels=data.keys(), # 设置饼图标签
            # colors=["#d5695d", "#5d8ca8", "#65a479", "#a564c9"], # 设置饼图颜色
            autopct="%.2f%%"
        )
    plt.title(dis_type) # 设置标题
    plt.show()
```

```python

for dis in four_disease:
    temp_dict = {}
    for p in path_dict[dis]:
        *pref,distype, direct,  num = p.split('_')
        temp_dict['_'.join(pref)] = num.split('-')[-1].split('.')[0].replace("(", "").replace(")", "")
    draw_pie(temp_dict, dis)
```
